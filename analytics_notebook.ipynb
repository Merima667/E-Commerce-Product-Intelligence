```python
import numpy as np

ratings = np.array([4.5, 3.2, 5.0, 2.8, 4.1, 3.9])
price_placeholder = np.zeros(6)
category_weights = np.ones((3, 4))
years = np.arange(2010, 2025)
score_buffer = np.empty((2, 3))
print("=== Array Info ===")
arrays = {
        'ratings': ratings,
        'price_placeholder': price_placeholder,
        'category_weights': category_weights,
        'years': years,
        'score_buffer': score_buffer,
    }
for name, arr in arrays.items():
        print(f'  {name:<22s} shape={str(arr.shape):<12} dtype={arr.dtype}  ndim={arr.ndim}  size={arr.size}')

ratings_arr = np.array([4.5, 3.2, 5.0, 2.8, 4.1, 3.9, 4.7, 2.5, 5.0, 3.6])
review_count = np.array([120, 45, 200, 30, 89, 67, 150, 25, 310, 55])

normalized = ratings_arr * 10
weighted   = ratings_arr * np.log(review_count)
high_rated = ratings_arr > 4.0
quality    = (ratings_arr > 3.5) & (review_count > 100)

print("\n=== Vectorized Operations ===")
print(f" mean : {ratings_arr.mean():.2f}")
print(f" std  : {ratings_arr.std():.2f}")
print(f" max  : {ratings_arr.max():.2f}")

print("\n=== Axis Reductions ===")
matrix = np.vstack([ratings_arr, review_count])
print(f" col_means: {matrix.mean(axis=0).round(2)}")
print(f" row_means: {matrix.mean(axis=1).round(2)}")

print("\n=== Broadcasting (normalized) ===")
min_v, max_v = ratings_arr.min(), ratings_arr.max()
normalized = (ratings_arr - min_v) / (max_v - min_v)
print(f" {normalized.round(3)}")
```

```python
# Part 7
print("=== Missing Values ===")
missing = df.isna().sum()
missing_pct = (missing / len(df) * 100).round(1)
missing_df = pd.DataFrame({'missing': missing, 'pct': missing_pct})
missing_df = missing_df[missing_df['missing'] > 0].sort_values('missing', ascending=False)
print(missing_df)
```

```python
#Part 7 - Heatmap

import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')

# Uzmi samo kolone sa missing vrijednostima
cols_with_missing = missing_df.index.tolist()[:20]  # top 20
df_missing = df[cols_with_missing].isna()

fig, ax = plt.subplots(figsize=(14, 8))
im = ax.imshow(df_missing.T.values, aspect='auto', cmap='RdYlGn_r')
ax.set_yticks(range(len(cols_with_missing)))
ax.set_yticklabels(cols_with_missing, fontsize=8)
ax.set_xlabel('Row index')
ax.set_title('Missing Value Heatmap (red = missing)')
plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.savefig('data/processed/analytics/missing_heatmap.png', dpi=120, bbox_inches='tight')
print("Heatmap saved!")
```
